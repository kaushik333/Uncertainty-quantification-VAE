```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
train_var = pd.read_csv("z_train.csv")
test_var = pd.read_csv("z_test.csv")
rand_var = pd.read_csv("z_rand.csv")
```

```python
train_imgs = np.load("z_train_images.npz")
test_imgs = np.load("z_test_images.npz")
rand_imgs = np.load("z_rand_images.npz")
```

```python
plt.figure()
plt.title("Train set variances")
_ = plt.hist(train_var["variance"], bins=20)

plt.figure()
plt.title("Test set variances")
_ = plt.hist(test_var["variance"], bins=20)

plt.figure()
plt.title("Random-sample variances")
_ = plt.hist(rand_var["variance"], bins=20)

train_thresh = np.percentile(train_var["variance"], 99)
test_thresh = np.percentile(test_var["variance"], 99)
rand_thresh = np.percentile(rand_var["variance"], 99)
```

## Show uncertain images in train set

```python
train_idxs = list(train_var.loc[train_var["variance"]>train_thresh].index)
idx = 0
for i in train_idxs:
    if idx > 10:
        break
    plt.figure()
    plt.imshow(train_imgs["arr_0"][i,:][0])
    idx += 1
```

```python
test_idxs = list(test_var.loc[test_var["variance"]>test_thresh].index)
idx = 0
for i in test_idxs:
    if idx > 10:
        break
    plt.figure()
    plt.imshow(test_imgs["arr_0"][i,:][0])
    idx += 1
```

```python
rand_idxs = list(rand_var.loc[rand_var["variance"]>rand_thresh].index)
idx = 0
for i in rand_idxs:

    if idx > 10:
        break
    plt.figure()
    plt.imshow(rand_imgs["arr_0"][i,:][0])
    idx += 1
```
